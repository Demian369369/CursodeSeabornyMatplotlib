```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['a','b','c'],y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark',palette='dark', font='Verdana',font_scale=1)
sns.barplot(x=['a','b','c'],y=[1,3,2])
plt.show()
```
